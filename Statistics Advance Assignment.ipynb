**How the parameters of the normal distribution relate to the shape of the distribution:**

* The mean (μ) of the normal distribution is the center of the distribution.
* The standard deviation (σ) of the normal distribution controls the spread of the distribution. A larger standard deviation results in a wider distribution.
* The skewness of the normal distribution is determined by the third moment (γ1). A positive skewness indicates that the distribution is skewed to the right, while a negative skewness indicates that the distribution is skewed to the left.
* The kurtosis of the normal distribution is determined by the fourth moment (γ2). A positive kurtosis indicates that the distribution is more peaked than a normal distribution, while a negative kurtosis indicates that the distribution is less peaked than a normal distribution.

**Q8: What is the z score? State the importance of the z score.**

**Z-score:**

The z-score is a measure of how far a data point is from the mean of a distribution, in terms of standard deviations. It is calculated by subtracting the mean from the data point and then dividing by the standard deviation.

**Importance of the z-score:**

The z-score is a useful tool for comparing data points from different distributions. For example, if we have two students who take different tests, we can use the z-score to compare their scores and see which student did better relative to their classmates.

The z-score is also used in hypothesis testing to determine whether a difference between two groups is statistically significant.

**Example:**

Suppose we have a class of 30 students and the mean score on a test is 80. One student, John, scored 90 on the test. John's z-score is:

(90 - 80) / 10 = 1

This means that John's score is 1 standard deviation above the mean.

Another student, Mary, scored 70 on the test. Mary's z-score is:

(70 - 80) / 10 = -1

This means that Mary's score is 1 standard deviation below the mean.


**Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.**

**Central Limit Theorem:**

The central limit theorem (CLT) states that the distribution of sample means from a population with any distribution (not necessarily normal) will approach a normal distribution as the sample size increases, regardless of the shape of the population distribution.

**Significance of the Central Limit Theorem:**

The CLT is significant because it allows us to make inferences about a population based on a sample, even if the population distribution is not known. This is because the distribution of sample means will be approximately normal, regardless of the shape of the population distribution.

The CLT is used in a wide variety of statistical applications, such as:

* Hypothesis testing
* Confidence intervals
* Regression analysis
* Quality control
* Risk assessment

**Example:**

Suppose we have a population of heights that is normally distributed with a mean of 68 inches and a standard deviation of 2 inches. If we take a sample of 100 people from this population, the distribution of the sample mean will also be approximately normal, with a mean of 68 inches and a standard deviation of 2/sqrt(100) = 0.2 inches.

This means that we can use the normal distribution to make inferences about the population mean height, even though we do not know the exact shape of the population distribution.

* The sample size is large enough (n > 30).
* The population distribution is not significantly skewed or kurtosis.
* The observations are independent of each other.
